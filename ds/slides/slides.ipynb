## ANTENATAL COVERAGE

```python
import pandas as pd
import numpy as np
OUT = '/Users/edinhamzic/Symphony/wb_bangladesh/Bangladesh/output/all/'
```

```python
ANC2011 = '/Users/edinhamzic/Symphony/wb_bangladesh/Bangladesh/output/all/clusters_antenatal_coverage_union_2011.csv'
ANC2016= '/Users/edinhamzic/Symphony/wb_bangladesh/Bangladesh/output/all/clusters_antenatal_coverage_union_2016.csv'
```

```python
anc2011 = pd.read_csv(ANC2011)
anc2016 = pd.read_csv(ANC2016)
```

```python
anc2011 = anc2011[anc2011['DistrictName'] == 'Khulna ']
anc2016 = anc2016[anc2016['DistrictName'] == 'Khulna ']
```

```python
difference = (anc2016.iloc[:,2:] - anc2011.iloc[:,2:])/anc2011.iloc[:,2:]*100
difference = difference.transpose().sort_values(by=27)
difference = pd.DataFrame(difference)
difference['abs'] = np.absolute(difference[27])
difference = difference[difference['abs'] > 50]
remove_list = []
for value in difference['abs'].astype(str):
    if value == 'inf':
        remove_list.append(False)
    else:
        remove_list.append(True)
difference = difference.loc[remove_list,:]        
difference.to_csv(OUT + 'AntenataleCoverage_Khulna.csv')
```

##  Antenatale coverage
- We will try to provide a very plasti example of what machine learning can provide us
- In the case of antenatale coverage. We took the example of Kuhlna district which moved from overall good performers to overall low performers when we consider antenatale coverage.
- Now, to illustrate how the clustering actually works and what is going on whit Kuhlna district we want to show you how clustering did the job and what happened with Khulna over this time
- What we observed is following:
- Rate of distributed ECPs, Rate of performed permanant method (total) out of 15-49 population decreases considerably
- On the other side, Neonatal case fatality rate (EmOC), proportion of women population as well as proportion of females as head of household tripled
- This tells us that in case of Khulna even thouh some of the services increase considerably, such as proportion of registered births. We see that here due to the pure socio-demographic trends general service delivery is lagging.


## UNMET NEED

```python
import pandas as pd
import numpy as np

UNMET2011 = '/Users/edinhamzic/Symphony/wb_bangladesh/Bangladesh/output/all/clusters_unmet_needs_vars_union_2011.csv'
UNMET2016 = '/Users/edinhamzic/Symphony/wb_bangladesh/Bangladesh/output/all/clusters_unmet_needs_vars_union_2016.csv'


unmet2011 = pd.read_csv(UNMET2011)
unmet2016 = pd.read_csv(UNMET2016)

unmet2011 = unmet2011[unmet2011['DistrictName'] == 'Habiganj ']
unmet2016 = unmet2016[unmet2016['DistrictName'] == 'Habiganj ']

difference = (unmet2016.iloc[:,2:] - unmet2011.iloc[:,2:])/unmet2011.iloc[:,2:]*100
difference = difference.transpose().sort_values(by=20)
difference = pd.DataFrame(difference)
difference['abs'] = np.absolute(difference[20])
difference = difference[difference['abs'] > 50]
remove_list = []
for value in difference['abs'].astype(str):
    if value == 'inf':
        remove_list.append(False)
    else:
        remove_list.append(True)
difference = difference.loc[remove_list,:]        
display(difference)
difference.to_csv(OUT + 'UnmetNeedForFamilyPlanning_Habiganj.csv')
```
