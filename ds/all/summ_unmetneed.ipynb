# Tables: Unmet needs

```python
import pandas as pd
```

```python
DATA = '/Users/edinhamzic/Symphony/wb_bangladesh/Bangladesh/output/all/'
clusters_intersect_all = 'clusters_unmet_needs_vars_intersect_all.csv'
clusters_intersect_2016 = 'clusters_unmet_needs_vars_intersect_2016.csv'
clusters_intersect_2011 = 'clusters_unmet_needs_vars_intersect_2011.csv'
clusters_union_all = 'clusters_unmet_needs_vars_union_all.csv'
clusters_union_2016 = 'clusters_unmet_needs_vars_union_2016.csv'
clusters_union_2011 = 'clusters_unmet_needs_vars_union_2011.csv'
corr_2016 = 'corr_unmet_needs_2016.csv'
cca_2016 = 'cca_unmet_needs_2016.csv'
corr_2011 = 'corr_unmet_needs_2011.csv'
cca_2011 = 'cca_unmet_needs_2011.csv'
corr_cca = [corr_2011, cca_2011, corr_2016, cca_2016]
```

```python
def read_corr_cca(dir_path, files_list):
    data_dict = {}
    for file in files_list:
        data_dict[file] = pd.read_csv(dir_path+file)
    return data_dict
```

## Summary tables

```python
clusters_intersect = pd.read_csv(DATA+clusters_intersect_all)
clusters_union = pd.read_csv(DATA+clusters_union_all)
```

```python
clusters_intersect_summ = clusters_intersect.groupby(['cluster', 'year']).median().transpose()
clusters_union_summ = clusters_union.groupby(['cluster', 'year']).median().transpose()
```

```python
display(clusters_intersect_summ.head())
clusters_intersect_summ.to_csv(DATA+'/unmetneeds_summaries/unmetneeds_intersect.csv')
```

```python
display(clusters_union_summ.head())
clusters_union_summ.to_csv(DATA+'/unmetneeds_summaries/unmetneeds_union.csv')
```
