```python
all_vars = maternal.merge(under5, how='outer', on='Input variable')
all_vars.head()
```

```python
all_vars = all_vars.merge(unmet, how='outer', on='Input variable')
all_vars.head()
```

```python
all_vars.head()
```

```python
all_vars.columns = ['Input variable', 'Maternal mortality rate', 'Under-5 mortality',
                    'Unmet need for family planning', 'Antenatal coverage']
```

```python
all_vars.head()
```

```python
print(all_vars['Maternal mortality rate'].value_counts())
print(all_vars['Under-5 mortality'].value_counts())
print(all_vars['Unmet need for family planning'].value_counts())
print(all_vars['Antenatal coverage'].value_counts())
```

```python
all_vars['Maternal mortality rate'] = np.where(all_vars['Maternal mortality rate']=='Maternal mortality rate', 'Yes', 'No')
all_vars['Under-5 mortality'] = np.where(all_vars['Under-5 mortality']=='Under-5 mortality', 'Yes', 'No')
all_vars['Unmet need for family planning'] = np.where(all_vars['Unmet need for family planning']=='Unmet need for family planning', 'Yes', 'No')
all_vars['Antenatal coverage'] = np.where(all_vars['Antenatal coverage']=='Antenatal coverage', 'Yes', 'No')
all_vars.head()
```
