```python
program_names = {'C01': 'Area, Households, Population, Density by Residence and Community',
                 'C02': 'Distribution of Households and Population by Sex , Sex Ratio, Residence and Community',
                 'C03': 'Percentage Distribution of General Households by Size, Average size, Residence and Community', 
                 'C04': 'Percentage Distribution of Population by Age groups, Residence and Community',
                 'C05': 'Percentage Distribution of Population aged 10 years and above by Sex, Marital status, Residence and Community',
                 'C06': 'Distribution of Population aged 7 years and above by Literacy, Sex, Residence and Community',
                 'C07': 'Distribution of Population aged 3-14 years by Age groups, School attendance, Sex, Residence and Community',
                 'C08': 'Distribution of Population aged 15-29 years by Age groups, School attendance, Sex, Residence and Community',
                 'C09': 'Percentage Distribution of Population by Type of disability, Residence and Community',
                 'C10': 'Distribution of Population aged 7 years and above not attending school by Employment Status, Sex, Residence and Community',
                 'C11': 'Distribution of Population aged 7 years and above not attending school but Employed by Field of Activity, Sex, Residence and Community',
                 'C12': 'Distribution of Ethnic Households, Population by Sex, Residence and Community',
                 'C13': 'Distribution of Population by Religion, Residence and Community',
                 'C14': 'Percentage Distribution of General Households by Type of Structure,Toilet Facility, Residence and Community',
                 'C15': 'Distribution of General Households by Source of Drinking Water, Electricity Connection and Housing Tenancy Status, by Residence and Community',
                }
```

```python
DATA = '../../output/bbs/data/census2011/'
os.listdir(DATA)
```

```python
def create_census2011_metadata(path, pattern, out_path, desc_dict):
    datasets_dict = {'variable':[], 'census_program_code': [], 'census_program_definition': []}
    for path, subfolds, files in os.walk(path):
        for file in files:
            if bool(re.search(r'C[0-9][0-9].',file)):
                tmp = pd.read_csv(os.path.join(path,file), low_memory=False)
                datasets_dict['variable'] = datasets_dict['variable'] + list(tmp.columns)
                program_code = file.replace(".csv", "")[-3:]
                datasets_dict['census_program_code'] = datasets_dict['census_program_code'] + [program_code] * len(tmp.columns) 
                datasets_dict['census_program_definition'] = datasets_dict['census_program_definition'] + [desc_dict[program_code]] * len(tmp.columns)
    out = pd.DataFrame.from_dict(datasets_dict)
    out = out.drop_duplicates().sort_values(by='census_program_code')
    display(out)
    out.to_csv(os.path.join(out_path, 'census2011_metadata.csv'), index=False, index_label=False) 

                
    
```

```python
create_census2011_metadata(path=DATA, pattern='.csv', out_path=DATA, desc_dict=program_names)
```
