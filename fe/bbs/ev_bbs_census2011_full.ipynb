```python
def read_dbf(path):
    pattern = "/*.DBF"
    dbf_files = glob.glob(path + pattern)
    zila = list()
    zila_c = list()
    for file in dbf_files:
        print("#"*60)
        print(file)
        tmp = DBF(file)
        tmp = pd.DataFrame(iter(tmp))
        zila.append(tmp['ZILA'].unique())
        zila_c.append(len(tmp['ZILA'].unique()))
        display(tmp.head())
        print(tmp.shape)
        print(round(sys.getsizeof(tmp)/1000000,2))
        del tmp
    return zila,zila_c 
```
