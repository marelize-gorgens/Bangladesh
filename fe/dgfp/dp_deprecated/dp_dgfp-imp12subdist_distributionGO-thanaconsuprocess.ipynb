```python
import os
import glob
import collections
import pandas as pd
import numpy as np
```

```python
DATA = '../../data/dgfp/data/importance12_subdistricts/distributionGO_thanaconsuprocess/'
```

```python
def read_data(path, ext):
    files_list = glob.glob(path + '/*.' + ext)
    data = list()
    for file in files_list:
        print(file)
        file_name = os.path.split(file)[1][:-4]
        program = file.split("/")[-2]
        tmp = pd.read_csv(file, sep='\t')
        file_name_comp = file_name.split('-')
        tmp['Year'] = file_name_comp[0]
        tmp['Month'] = file_name_comp[1]
        tmp['District'] =file_name_comp[3]
        tmp['Division'] = file_name_comp[4]
        tmp['program'] = program
        data.append(tmp)
    return pd.concat(data, axis=0)
    
```

```python
data = read_data(path=DATA, ext='csv')
```
